## **Data Cleaning and Filtering**
<a href=#cont>Back to Table of Contents</a>

* **Purpose:** Prepare the data for analysis by cleaning and filtering.
* **Details:** Include steps for handling missing values, removing outliers, correcting errors, and possibly reducing the data (filtering based on certain criteria or features).
---

```python
# Dataframe 1 for data cleanup
df1 = df

# Drop duplicates if any
df1 = df1.drop_duplicates()

# Handling missing values
# Use SimpleImputer to fill missing numeric values with the mean
imputer = SimpleImputer(strategy='mean')

# Columns with missing values
missing_columns = ['RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)', 'RICE YIELD (Kg per ha)',
                   'WHEAT AREA (1000 ha)', 'WHEAT PRODUCTION (1000 tons)', 'WHEAT YIELD (Kg per ha)',
                   'SUGARCANE AREA (1000 ha)', 'SUGARCANE PRODUCTION (1000 tons)', 'SUGARCANE YIELD (Kg per ha)',
                   'COTTON AREA (1000 ha)', 'COTTON PRODUCTION (1000 tons)']

df1[missing_columns] = imputer.fit_transform(df1[missing_columns])

# Filter out unrealistic values (outliers)
# Filtering out negative values and extremely high values

# Filter for reasonable crop area values (e.g., between 0 and 10000 ha)
df1 = df1[(df1['RICE AREA (1000 ha)'] >= 0) & (df1['RICE AREA (1000 ha)'] <= 10000)]
df1 = df1[(df1['WHEAT AREA (1000 ha)'] >= 0) & (df1['WHEAT AREA (1000 ha)'] <= 10000)]
df1 = df1[(df1['SUGARCANE AREA (1000 ha)'] >= 0) & (df1['SUGARCANE AREA (1000 ha)'] <= 10000)]
df1 = df1[(df1['COTTON AREA (1000 ha)'] >= 0) & (df1['COTTON AREA (1000 ha)'] <= 10000)]

# Filter for realistic production values (e.g., between 0 and 50000 tons)
df1 = df1[(df1['RICE PRODUCTION (1000 tons)'] >= 0) & (df1['RICE PRODUCTION (1000 tons)'] <= 50000)]
df1 = df1[(df1['WHEAT PRODUCTION (1000 tons)'] >= 0) & (df1['WHEAT PRODUCTION (1000 tons)'] <= 50000)]
df1 = df1[(df1['SUGARCANE PRODUCTION (1000 tons)'] >= 0) & (df1['SUGARCANE PRODUCTION (1000 tons)'] <= 50000)]
df1 = df1[(df1['COTTON PRODUCTION (1000 tons)'] >= 0) & (df1['COTTON PRODUCTION (1000 tons)'] <= 50000)]

# Removing outliers using Z-score for 'YIELD' columns
yield_columns = ['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)', 'SUGARCANE YIELD (Kg per ha)', 'COTTON YIELD (Kg per ha)']
for column in yield_columns:
    z_scores = np.abs(stats.zscore(df[column]))
    df1 = df1[z_scores < 3]  # Retain only rows with Z-score less than 3

# Dropping unnecessary columns
df1 = df1.drop(columns=['Dist Code', 'State Code'])

# Display the cleaned data
print(df1.info())
print(df1.describe())
```

# Data Cleaning: Standardizing Column Names and Identifying Missing Values

```python
# Handling column names for easier access
df1.columns = df1.columns.str.replace('\n', '').str.strip()

# Identify columns with non-numeric values
for column in df1.columns:
    try:
        df1[column].astype(float)
    except ValueError:
        print(f"Non-numeric values found in column: {column}")


# Check for missing values after conversion
print("Missing values after conversion:")
print(df1.isnull().sum())


# Display the cleaned data
print(df1.info())
print(df1.describe())
```

# Clustering-Based Anomaly Detection (DBSCAN):

```python
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df2[numerical_columns])

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)  # Adjust parameters as needed
clusters = dbscan.fit_predict(scaled_data)

# Add cluster labels to the dataframe
df2['Cluster'] = clusters

# Detect anomalies (label -1 in DBSCAN)
anomalies = df2[df2['Cluster'] == -1]

print(f"### Anomalies Detected Using DBSCAN: {len(anomalies)}")
print(anomalies)
```
